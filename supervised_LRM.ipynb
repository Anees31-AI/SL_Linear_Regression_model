**Linear Regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


```

**Sample Dataset**

```python
data={
    'StudyHours':[1, 2, 3, 4, 5, 6, 7, 8, 9],
    'AttendanceRate':[60, 65, 70, 75, 80, 85, 90, 92, 95],
    'PreviousScore':[30, 40, 50, 55, 60, 62, 68, 75, 80],
    'SleepHours':[6, 5.5, 6.5, 7, 7, 6, 6.5, 7.5, 8],
    'Scores':[35,40,50,55,60,65,70,78,85]
}
df=pd.DataFrame(data)
df.head()
```

**Training a Linear Regression**


*   Split Data




```python
X=df[['StudyHours','AttendanceRate','PreviousScore','SleepHours']]
y=df['Scores']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

**Train Model**

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

**Predict**

```python
y_pred=model.predict(X_test)
```

**Evaluate**

```python
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print("RMSE with multiple features:",rmse)
```

**Visualize the regression line**

```python
#Plot data points
plt.scatter(X['StudyHours'],y,color='blue',label='Actual')

#Regression Line
line=model.coef_* X+model.intercept_
plt.plot(X,line,color='red',label='Regression Line')

plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.title("Study Hours vs Exam score")
plt.legend()
plt.show()
```

**Make Predictions**

```python
new_student = pd.DataFrame({
    'StudyHours': [7.5],
    'AttendanceRate': [90],
    'PreviousScore': [70],
    'SleepHours': [6.5]
})

predicted_score = model.predict(new_student)
print(f"Predicted Score: {predicted_score[0]:.2f}")
```
